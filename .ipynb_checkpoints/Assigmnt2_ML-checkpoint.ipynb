```python
import cv2
import pandas as pd

img_path = r'aqua.jpg'
img = cv2.imread(img_path)

# Global variables declaration
clicked = False
r = g = b = x_pos = y_pos = 0

# Dataset reading
index = ["color", "color_name", "hex", "R", "G", "B"]
csv = pd.read_csv('colors.csv', names=index, header=None)

# Get x,y values of mouse click 
def draw_function(event, x, y, flags, param):
  
     if event == cv2.EVENT_LBUTTONDOWN:
        global b, g, r, x_pos, y_pos, clicked
        clicked = True
        x_pos = x
        y_pos = y
        b, g, r = img[y, x]
        b = int(b)
        g = int(g)
        r = int(r)
        
# Get the most matching color
def get_color_name(R, G, B):
    minimum = 10000
    
    for i in range(len(csv)):
        d = abs(R - int(csv.loc[i, "R"])) + abs(G - int(csv.loc[i, "G"])) + abs(B - int(csv.loc[i, "B"]))
        if d <= minimum:
            minimum = d
            cname = csv.loc[i, "color_name"]
    return cname

# Get the hexadecimal value of the selected colour
def get_hex(R, G, B):
    minimum = 10000
    
    for i in range(len(csv)):
        d = abs(R - int(csv.loc[i, "R"])) + abs(G - int(csv.loc[i, "G"])) + abs(B - int(csv.loc[i, "B"]))
        if d <= minimum:
            minimum = d
            hex_ = csv.loc[i, "hex"]
    return hex_

cv2.namedWindow('image')
cv2.setMouseCallback('image', draw_function)

while True:

    cv2.imshow("image", img)
    if clicked:

        cv2.rectangle(img, (20, 20), (900, 60), (b, g, r), -1)

        # String to display rgb values and names of the colours
        text =get_color_name(r, g, b) + ' R = ' + str(r) + ' G = ' + str(g) + ' B = ' + str(b) + ' Hex = ' + get_hex(r, g, b)
        cv2.putText(img, text, (50, 50), 2, 0.8, (255, 255, 255), 2, cv2.LINE_AA)

        # Using black text for lighter colours
        if r + g + b >= 600:
            cv2.putText(img, text, (50, 50), 2, 0.8, (0, 0, 0), 2, cv2.LINE_AA)

        clicked = False

    # exit window using "esc" key
    if cv2.waitKey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
